*Writing code to practice **series and dataframes** for data handling*

```python
# series - 1D representations
import pandas as pd

book= ["dhina", "AK", "red dragon", "billa"]
series=pd.Series(book)
print(series)
```

```python
book={
 "dhina": 24, "AK": 53, "red dragon": 53, "billa": 30
}

series=pd.Series(book)
print(series)
```

```python
print(series[series>30])
```

```python
# 2D - dataframes
import pandas as pd

books={
    "hero": ["dhina", "AK", "red dragon"],
    "villain": ["jesse", "james", "meowth"]
}

df = pd.DataFrame(books)
print(df)
```

```python
# adding a new column
df["hero powers"]=["chaining", "patience", "machine gun"]
df["villain powers"]=["rocket", "rocket", "mimick"]

print(df)
```

```python
# adding a new row
add_row=pd.DataFrame([{
    "hero": "billa",
    "villain": "dimitiri",
    "hero powers": "head of criminals",
    "villain powers": "gun alliance"
}])

df=pd.concat([df, add_row], ignore_index=True)
print(df)
```

```python
# remove duplicate rows
df=df.drop_duplicates()
print(df)
```

```python
print("\n",df.head(2))
print("\n",df.tail(2))
print("\nshape:",df.shape)
```
