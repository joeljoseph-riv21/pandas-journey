*Writing code to practice **EDA and data cleaning** by working with dropna()& fillna() methods*

```python
import kagglehub
path = kagglehub.dataset_download("benroshan/factors-affecting-campus-placement")
```

```python
import os
for root, _, filename in os.walk(path):
    for files in filename:
        print(f"filename:{files}")

        file_path= os.path.join(root, files)
        print(f"file_path:{file_path}")
```

```python
import pandas as pd
df= pd.read_csv(file_path)
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print(df.describe(include='all'))
```

```python
print(df.isnull().sum())
```

```python
groups= df.groupby(['salary'])
grouped= groups.min()
print(grouped.iloc[0:13,9:15])
```

```python
# fill missing 'salary' values with 0
df_zero = df.copy()
df_zero['salary'] = df_zero['salary'].fillna(0)
print(df_zero[df_zero['status'] == 'Not Placed'].head())
```

```python
# fill missing 'salary' values with the mean
# calculate the mean salary of placed students
mean_salary= df[df['status'] == 'Placed']['salary'].mean()

df_mean= df.copy()
df_mean['salary']= df_mean['salary'].fillna(mean_salary)
print(df_mean[df_mean['status'] == 'Not Placed'].head())
```

```python
# drop rows where 'salary' is NaN
df_drop= df.copy()
df_drop= df_drop.dropna(subset=['salary'])

print(f"original number of rows: {df.shape[0]}")
print(f"number of rows after dropping: {df_drop.shape[0]}")
```
