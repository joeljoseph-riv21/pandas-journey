*Writing code to practice **exploratory data analysis (EDA)** covering the  foundational EDA practices*

```python
# downloading a dataset from kaggle
import kagglehub
path=kagglehub.dataset_download("lokatiraghuvaran/cafe-sales-dirty-and-clean-dataset")
```

```python
print(path)
```

```python
# fixing the complex nested directory structure issue by os.walk()
import os
import pandas as pd
for root, _, filenames in os.walk(path):
    for filename in filenames:
        print(f"\nfiles:{filenames}")
        print(os.path.join(root, filename))

print(f"\nroot dir path: {root}")
```

```python
# inspecting the cleaned dataset
clean_path=os.path.join(path, 'data', 'processed', 'cafe_sales_cleaned.csv')
df=pd.read_csv(clean_path)

print(f"loaded data from:{clean_path}")
print(f"shape:{df.shape}")
display(df.head())
```

```python
# exploring and cleaning practice on raw dataset
dirt_path=os.path.join(path, 'data', 'raw', 'cafe_sales_dirty.csv')
df=pd.read_csv(dirt_path)

print(f"loaded data from:{dirt_path}")
print(f"shape:{df.shape}")
display(df.head())
```

```python
# getting basic information on rows, columns and types
print(df.info())
```

```python
# getting statistical summary on all numeric& categorical columns
print(df.describe(include='all'))
```

```python
# we know there are null values, summing up them here
print(df.isnull().sum())
```

```python
# we check if any duplicates, print their count
"""
- df.duplicated() returns a bool Series list marking True for duplicated rows,
passing it inside [] pandas uses this Series to select rows.

- Where, the True marked rows from df.duplicated() gets filtered by using
df[df.duplicated()] from the original df and returns a new DataFrame
containing only the duplicate rows.

- This common pandas technique is called boolean indexing.
"""

duplicate_rows=df[df.duplicated()]
print(f"number of duplicate rows: {duplicate_rows.shape[0]}")

if not duplicate_rows.empty:
    print("duplicate rows:")
    display(duplicate_rows)
else:
    print("\nno duplicates.")
```

```python
# check for duplicate column names
duplicate_columns=df.columns[df.columns.duplicated()]

if not duplicate_columns.empty:
    print(f"duplicate columns: {list(duplicate_columns)}")
else:
    print("no duplicate columns.")
```
