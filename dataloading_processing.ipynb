*Writing code to practice **data loading, dataframe indexing and slicing, aggregation functions, grouping and filtering data** in pandas*

```python
# reading from the csv file
import pandas as pd

df=pd.read_csv("/content/sample_data/california_housing_test.csv")
print(df.columns)
print(df.shape)
```

```python
# getting information about the dataframe
print(df.info())
```

```python
# printing the first 5 rows for inspection
print(df.head(5))
```

```python
# accessing the rows and columns with indexing& slicing
print(df.iloc[0:10, 0:3])
```

```python
# selecting a column and picking the minimum value
print(df["total_rooms"].min())
```

```python
# grouping dataframes using groupby() and use by applying aggregated function
grouping=df.groupby("population")
aggregated_df=grouping.mean()

aggregated_df=(aggregated_df.iloc[0:21, 0:3])
print(aggregated_df)
```

```python
# filtering the grouped dataframe
filtered_df = aggregated_df[aggregated_df["housing_median_age"] > 33]
print(filtered_df)
```
